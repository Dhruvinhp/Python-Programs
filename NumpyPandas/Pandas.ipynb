```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df.dropna() #drop NaN data columns
```

```python
df.dropna(how = 'all', axis=1) #if all column data is NaN then drop the column
```

```python
df.drop_duplicates(subset=['name'], keep='first') #drop duplicate values but keep first occured value column
```

```python
df.info()
```

```python
df['toy'].value_counts(dropna=True) #gives unique values, ignore NaN values.
```

```python
df.notnull() #check NaN values
```

```python
df.isnull()
```
