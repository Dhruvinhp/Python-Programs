```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
plt.plot([1,2,3],[2,5,7])
plt.title("Dhruvin's data")
plt.ylabel("y axis")
plt.xlabel("X axis")
plt.show()
```

```python
style.use('ggplot')
x = [1,5,9]
y = [24,45,38]
x2 = [4,6,8]
y2 = [1,2,9]
plt.scatter(x,y, label = 'first', linewidth=3)
plt.plot(x2,y2, label = 'second', linewidth = 2)
plt.hist(x, label = 'third', linewidth=3)
plt.title("Dhruvin's data")
plt.ylabel("y axis")
plt.xlabel("X axis")
plt.legend()
plt.show()
```
